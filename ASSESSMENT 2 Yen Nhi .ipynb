## 4. TIME SERIES

```python
filtered_df = df.loc[(df['Food Item'] == 'Bread') & (df['Country'] == 'Australia')]
filtered_df.head(12)
```

```python
# INSPECTION: Consider below
df.head(12)
```

```python
filtered_df = filtered_df.assign(MONTH=filtered_df['Month'].apply(lambda x: str(x).zfill(2)))
filtered_df.head(12)
```

```python
# FORMATTING: In a new DataFrame, create a new column, 'MONTH' based on the original 'Month' column
# Hint: Use .assign() to add leading zeros in the month column. Note: the datatype becomes string

```

```python
new_df = df.assign(DATE=df['Year'].astype(str) + '-' + df['Month'].astype(str).str.zfill(2) + '-01')
new_df.head(12)
```

```python
# FORMATTING: In a new DataFrame, create a new column, 'DATE' to hold a time series
# Hint: Use .assign() to combine values from two columns into strings

# add the 01 to date to make it the first day of the month
```

```python
new_df['DATE'] = pd.to_datetime(new_df['DATE'])
new_df.head(12)
```

```python
# DATATYPE CONVERSION: Convert the column, 'DATA' into the datatype, datetime64[ns]
# Hint: Use .to_datetime() to combine string dates to datatime date

```

```python
new_df.info()
```

```python
# Datatype check. Notice the new column, DATE is not datetime64[ns]. Ready for time series!
```

```python
overall_avg = new_df.pivot_table(index='DATE', columns='Country', values='Price in USD', aggfunc='mean')
overall_avg.head()
```

```python
# Time Series: Display and plot overall average food prices of these five countries of each month from 2018-2022
# Hint: Step 1: Use .pivot_table() to convert dates into index and calculate aggregated mean values
# Hint: Step 2: Plot the above values into a bar( i think it is line graph) chart (see below)

```

```python
overall_avg.plot(kind='line', figsize=(12, 6))

plt.xlabel('DATE')
plt.ylabel('USD')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1,1), loc ='upper left')
plt.tight_layout()
plt.show()
```

```python
# Time Series: Display and plot overall average food prices of these five countries of each month from 2018-2022
# Hint: Step 2: Plot the above values into a bar chart (see below)

# looking at the graph the overal average food prices is growing over the years, and Australia has the steepest growth since 2021.
```

```python
australia_df = new_df[(new_df['Country'] == 'Australia') & (new_df['Year'].isin([2020, 2021]))]
australia_prices = australia_df.pivot_table(index='DATE', columns='Year', values='Price in USD', aggfunc='mean')

australia_prices = australia_prices.resample('ME').mean()


australia_prices_shifted = australia_prices.copy()
australia_prices_shifted[2020] = australia_prices_shifted[2020].shift(12)
australia_prices_shifted = australia_prices_shifted.loc['2021-01':]

australia_prices_shifted.plot();

```

```python
# Compare the average food prices in Australia between two years (2020 vs 2021) for the period of 12 months via a plot (see below)

# i can see that in year 2021 food price increase really fast compare to growth rate of 2020 food price in Australia
```

```python
japan_df = new_df[(new_df['Country'] == 'Japan') & (new_df['Year'].isin([2020, 2021]))]
japan_prices = japan_df.pivot_table(index='DATE', columns='Year', values='Price in USD', aggfunc='mean')

japan_prices = japan_prices.resample('ME').mean()

japan_prices_shifted = japan_prices.copy()
japan_prices_shifted[2020] = japan_prices_shifted[2020].shift(12)
japan_prices_shifted = japan_prices_shifted.loc['2021-01':]

japan_prices_shifted.plot();


```

```python
# Compare the average food prices in Japan between two years (2020 vs 2021) for the period of 12 months via a plot (see below)

# food price in japan is also increasing
```
